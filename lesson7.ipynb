```python
r0 = plasma['r0'][0]
rm = plasma['rm'][0]
z0 = plasma['z0'][0]
r0, rm, z0
```

```python
import numpy as np

def mag_surf(th, xr, xdl, xly, xgm):
    cotet=np.cos(th)
    sitet=np.sin(th)
    xx=-xdl+xr*cotet-xgm*sitet**2
    zz=xr*xly*sitet
    x=(r0+rm*xx)/100
    z=(z0+rm*zz)/100
    return x,z
```
